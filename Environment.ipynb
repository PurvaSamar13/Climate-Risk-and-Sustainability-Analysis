```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("per-capita-co-emissions.csv")
data
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)

# Sorting the DataFrame by Annual CO2 emissions (per capita) in descending order
df_sorted = df.sort_values(by='Annual CO2 emissions (per capita)', ascending=False)

# Plotting the bar graph
plt.figure(figsize=(15, 7))
sns.barplot(x='Year', y='Annual CO2 emissions (per capita)', hue='Country', data=df_sorted, dodge=False)

plt.title('Annual CO2 Emissions (per capita): India, United States, and China (2013-2022)')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions (per capita)')
plt.xticks(rotation=0)
plt.legend(title='Country')
plt.tight_layout()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Prepare the data for regression
X = df[['Year']]
y = df['Annual CO2 emissions (per capita)']

# Initialize a dictionary to hold the predictions for each country
predictions_2025 = {}

# Loop through each country and fit a linear regression model
for country in df['Country'].unique():
    # Filter the data for the current country
    df_country = df[df['Country'] == country]
    X_country = df_country[['Year']]
    y_country = df_country['Annual CO2 emissions (per capita)']
    
    # Create and fit the model
    model = LinearRegression()
    model.fit(X_country, y_country)
    
    # Predict for 2025
    prediction = model.predict(np.array([[2025]]))
    
    # Store the prediction
    predictions_2025[country] = prediction[0]

# Print the predictions for 2025
print("Predicted CO2 Emissions for 2025 (per capita):")
for country, emission in predictions_2025.items():
    print(f"{country}: {emission:.2f}")
```

```python
# Function to prepare the data
def prepare_data(df, countries, indicator):
    # Filter data for the specified countries and indicator
    df_filtered = df[(df['Country'].isin(countries)) & (df['Indicator'] == indicator)]
    if df_filtered.empty:
        print("No data available for the specified countries or indicator.")
        return None
    
    years = [str(year) for year in range(2013, 2023)]
    df_filtered = df_filtered[['Country', 'Indicator'] + years]
    
    # Melt the DataFrame
    df_melted = pd.melt(df_filtered, id_vars=['Country', 'Indicator'], value_vars=years, var_name='Year', value_name='Value')
    df_melted['Year'] = df_melted['Year'].astype(int)
    return df_melted

# Function to predict values
def predict_values(df_melted, countries):
    predictions = {}
    for country in countries:
        country_data = df_melted[df_melted['Country'] == country]
        X = country_data['Year'].values.reshape(-1, 1)
        y = country_data['Value'].values
        
        model = LinearRegression()
        model.fit(X, y)
        
        year_2025 = np.array([[2025]])
        prediction = model.predict(year_2025)
        predictions[country] = prediction[0]
    
    return predictions

# List of countries and the indicator
countries = ['United States', 'China, P.R.: Mainland', 'India']
indicator = 'Climate-driven Hazard & Exposure'  # Adjust based on actual indicator names

# Assuming 'df' is your DataFrame
# Prepare the data
df_melted = prepare_data(df, countries, indicator)

# Make predictions
if df_melted is not None:
    predictions = predict_values(df_melted, countries)
    print("Predicted Climate-driven Hazard & Exposure for 2025:")
    for country, value in predictions.items():
        print(f"{country}: {value:.2f}")
```

```python
data = pd.read_excel("Climate_Risk.xlsx",sheet_name = "Sheet1")
data
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

# Convert to DataFrame
df = pd.DataFrame(data)

# Prepare data for each indicator
indicators = df['Indicator'].unique()
train_size = 8
test_size = 2
overall_mape = []

# Evaluate each indicator
for indicator in indicators:
    df_indicator = df[df['Indicator'] == indicator].drop(columns=['Country', 'Indicator'])
    values = df_indicator.values.flatten()
    train, test = values[:train_size], values[train_size:train_size + test_size]
    years = pd.date_range(start='2013', periods=len(train), freq='Y')
    
    # Fit the model on training data
    df_train = pd.DataFrame(train, index=years, columns=[indicator])
    model = ARIMA(df_train, order=(1, 0, 0))
    model_fit = model.fit()
    
    # Predict on test data
    forecast = model_fit.forecast(steps=test_size)
    
    # Evaluate the model's MAPE
    mape = mean_absolute_percentage_error(test, forecast)
    overall_mape.append(mape)
    
    # Calculate accuracy
    accuracy = 100 - mape
    
    # Print the results
    print(f"{indicator} - Test Set Evaluation")
    print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
    print(f"Accuracy: {accuracy:.2f}%\n")

# Calculate overall accuracy
overall_accuracy = 100 - np.mean(overall_mape)
print(f"Overall Model Accuracy: {overall_accuracy:.2f}%")
```
